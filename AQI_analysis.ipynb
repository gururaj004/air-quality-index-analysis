```python
# This handles different columns automatically
# Missing columns will be filled with NaN
df_merged = pd.concat([df1, df2, df3, df4, df5], 
                       axis=0,              # stack rows on top of each other
                       ignore_index=True,   # reset index
                       sort=False)          # keep original column order

print("Merged shape:", df_merged.shape)
print("\nAll columns after merge:")
print(df_merged.columns.tolist())
print("\nMissing values after merge:")
print(df_merged.isnull().sum())
```

```python
# Fill missing pollutant values with mean
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 
              'NH3', 'CO', 'SO2', 'O3', 
              'Benzene', 'Toluene', 'Xylene']

# Only fill columns that exist
existing_pollutants = [p for p in pollutants if p in df_merged.columns]

df_merged[existing_pollutants] = df_merged.groupby('City')[existing_pollutants].transform(
    lambda x: x.fillna(x.mean())
)

# Drop rows where AQI is missing
df_merged.dropna(subset=['AQI'], inplace=True)

print("\nFinal merged shape:", df_merged.shape)
```

```python
# Now use df instead of df_merged
df = df_merged.copy()
print("Ready to use! Shape:", df.shape)
```

```python
# Adding City and State to station_day files
df3 = pd.merge(df3, df5[['StationId', 'City', 'State']], 
               on='StationId', 
               how='left')

# Add City and State to station_hour
df4 = pd.merge(df4, df5[['StationId', 'City', 'State']], 
               on='StationId', 
               how='left')

print("Station day columns:", df3.columns.tolist())
print("Station hour columns:", df4.columns.tolist())
```

```python
# Rename Datetime to Date
df2 = df2.rename(columns={'Datetime': 'Date'})
df4 = df4.rename(columns={'Datetime': 'Date'})

print("Renamed successfully!")
```

```python
df3 = df3.drop(columns=['StationId'])
df4 = df4.drop(columns=['StationId'])

print("StationId dropped!")
```

```python
# Merge all 4 files into one
df_merged = pd.concat([df1, df2, df3, df4],
                       axis=0,
                       ignore_index=True,
                       sort=False)

print("Merged shape:", df_merged.shape)
print("\nColumns:", df_merged.columns.tolist())
```

```python
# Fix date column
df_merged['Date'] = pd.to_datetime(df_merged['Date'],infer_datetime_format=True,
                                    errors='coerce')

# Extract year month season
df_merged['Year'] = df_merged['Date'].dt.year
df_merged['Month'] = df_merged['Date'].dt.month

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Summer'
    elif month in [6, 7, 8, 9]:
        return 'Monsoon'
    else:
        return 'Post-Monsoon'

df_merged['Season'] = df_merged['Month'].apply(get_season)

# Fill missing pollutant values
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx',
              'NH3', 'CO', 'SO2', 'O3',
              'Benzene', 'Toluene', 'Xylene']

df_merged[pollutants] = df_merged.groupby('City')[pollutants].transform(
    lambda x: x.fillna(x.mean())
)

# Drop rows where AQI is missing
df_merged.dropna(subset=['AQI'], inplace=True)

print("Cleaned shape:", df_merged.shape)
```

```python
df = df_merged.copy()
print("Final shape:", df.shape)
df.head()
```

```python
# Check final dataset
print("Final Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nMissing Values:")
print(df.isnull().sum())
print("\nFirst 5 rows:")
df.head()
```

```python
print("Missing values before cleaning:")
print(df.isnull().sum())
```

```python
# Drop rows where City is completely missing
df.dropna(subset=['City'], inplace=True)
print("After dropping missing City rows:", df.shape)
```

```python
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx',
              'NH3', 'CO', 'SO2', 'O3',
              'Benzene', 'Toluene', 'Xylene']

# Only fill columns that exist
existing_pollutants = [p for p in pollutants if p in df.columns]

# Fill with city wise mean
df[existing_pollutants] = df.groupby('City')[existing_pollutants].transform(
    lambda x: x.fillna(x.mean())
)
print("Pollutant missing values filled!")
```

```python
# Fill missing State with Unknown
df['State'] = df['State'].fillna('Unknown')
print("State missing values filled!")
```

```python
# Fill AQI_Bucket based on AQI value
def get_aqi_bucket(aqi):
    if pd.isna(aqi):
        return 'Unknown'
    elif aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Satisfactory'
    elif aqi <= 200:
        return 'Moderate'
    elif aqi <= 300:
        return 'Poor'
    elif aqi <= 400:
        return 'Very Poor'
    else:
        return 'Severe'

df['AQI_Bucket'] = df.apply(
    lambda row: get_aqi_bucket(row['AQI']) 
    if pd.isna(row['AQI_Bucket']) 
    else row['AQI_Bucket'], axis=1)

print("AQI Bucket filled!")
```

```python
df.dropna(subset=['AQI'], inplace=True)
print("Dropped rows with missing AQI!")
```

```python
# Fill remaining numeric columns with mean
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
print("Remaining numeric missing values filled!")
```

```python
print("\nMissing values after cleaning:")
print(df.isnull().sum())
print("\nFinal shape:", df.shape)
```

```python
# Fill missing Date from Datetime and vice versa
df['Date'] = df['Date'].fillna(df['Datetime'])
df['Datetime'] = df['Datetime'].fillna(df['Date'])

# Convert to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Datetime'] = pd.to_datetime(df['Datetime'], errors='coerce')

print("Date and Datetime fixed!")
```

```python
# Fill missing StationId and StationName with Unknown
df['StationId'] = df['StationId'].fillna('Unknown')
df['StationName'] = df['StationName'].fillna('Unknown')

print("StationId and StationName fixed!")
```

```python
# Fill missing Status with Active
df['Status'] = df['Status'].fillna('Active')

print("Status fixed!")
```

```python
print("Missing values after fixing:")
print(df.isnull().sum())
print("\nFinal shape:", df.shape)
```
